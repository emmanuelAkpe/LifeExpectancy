#### Data cleaning

```python
column_mean=LifeExpectancyData.mean()
```

```python
# we fill the missing values with the mean of each column
LifeExpectancyData1= LifeExpectancyData.fillna(column_mean)
```

```python
LifeExpectancyData1.isna().sum()
```

####  Data Analysis 
We analysed our data with respect to Immunization factors, Economic factors, Health Issues, and Social factors and their relationship with Life Expectancy.

```python
#Displaying the general correlation between the variables in the dataset in tabular form
LifeExpectancyData1.corr()
```

```python
#Displaying the general correlation between the variables in the dataset in graphical form
import seaborn as sns
sns.heatmap(LifeExpectancyData1.corr())
```

```python

#plotting a scatter diagram to show the general trend of the Life Expectancy over time.
plt.scatter(LifeExpectancyData1['Year'], LifeExpectancyData1['Life expectancy '])
plt.xlabel('Years')
plt.ylabel('Life expectancy')
```

```python

LifeExpectancyData.columns
```

```python
from scipy.interpolate import *
import seaborn as sns
%matplotlib inline
```

```python

# Relatinship between Life expectancy and Immunization factors
sns.pairplot(LifeExpectancyData1,
            x_vars=['Hepatitis B','Measles ','Diphtheria ' ,'Polio' ], 
            y_vars=['Life expectancy '],
           kind='reg' )
```

```python
#Relationship between Economic factors and Life Expectancy
sns.pairplot(LifeExpectancyData1,
            x_vars=['percentage expenditure','GDP','Income composition of resources' ,'Total expenditure' ], 
            y_vars=['Life expectancy '],
           kind='reg' )
```

```python
# Relationship between Life expectancy and mortality
sns.pairplot(LifeExpectancyData1,
            x_vars=['Adult Mortality','infant deaths','under-five deaths ' ], 
            y_vars=['Life expectancy '],
           kind='reg' )
```

```python

# Relationship between Social factors and Life Expectancy
sns.pairplot(LifeExpectancyData1,
            x_vars=['Alcohol','Schooling' ], 
            y_vars=['Life expectancy '],
           kind='reg' )
```

```python
# Relationship between Health factors and Life Expectancy
sns.pairplot(LifeExpectancyData1,
            x_vars=[' BMI ',' HIV/AIDS',' thinness  1-19 years' ,' thinness 5-9 years' ], 
            y_vars=['Life expectancy '],
           kind='reg' )
```
